```python
@jit(nopython=True)
def hvovec(lon1=0.0, lat1=0.0, lon2=0.0, lat2=0.0, rad=False):

    '''
    Parameters
    ----------
    lon1 : float
        Longitude of first point.
    lat1 : float
        Latitude of first point.


    lon2 : float
        Longitude of second point.
    lat2 : float
        Latitude of second point.

    rad : Boolean, optional
        (default) False -> Angle is returned in Degrees
        True -> Angle is returned in radians
    

    Returns
    -------
    float
        Returns the haversine angle between two vectors given their latitude and longitude
    

    Warning
    -------
        This function assumes the input to be in degrees and of equal length\n
        or one of the input pairs to be a single value and the other pair to be an array
    '''

    #Convert decimal degrees to Radians:
    lon1 = np.deg2rad(lon1)
    lat1 = np.deg2rad(lat1)
    lon2 = np.deg2rad(lon2)
    lat2 = np.deg2rad(lat2)

    #Implementing Haversine Formula: 
    dlon = np.subtract(lon2, lon1)
    #dlat = np.subtract(lat2, lat1)

    a = np.add(np.multiply(np.sin(lat1), np.sin(lat2)), np.multiply(np.multiply(np.cos(lat1), np.cos(lat2)), np.cos(dlon)))

    if rad == True:
        return np.arccos(a)
    else:
        return np.abs(np.rad2deg(np.arccos(a)))
```

## SINGLE SOURCE ANALYSIS

```python
@jit(nopython=True)
def sig(psrno, declcut): 

    '''
    Parameters
    ----------
    psrno : int
        Pulsar number from the ATNF catalogue.
    
    declcut : float
        Declination cut-off in degrees.

    Returns
    -------
    Returns the signal PDF for the {psrno}th pulsar 
    and the all neutrinos within the specified declcut

    Also returns the number of neutrinos within the specified declcut
    '''
    decdf = np.abs(np.subtract(icdec, msdec[psrno])) #Declination difference between the pulsar and the neutrino
    nuind = np.nonzero(decdf < declcut) #indices of neutrinos within declination cut-off
    
    #Square of angular distance between the pulsar and the neutrino
    ang2 = hvovec(msra[psrno], msdec[psrno], icra[nuind], icdec[nuind], rad=True) ** 2 
    sg = np.deg2rad(icang[nuind]) ** 2 #Square of angular error of the neutrino
    return ((np.exp(-1 * ang2 / (2 * sg)) / (2 * np.pi * sg)), len(nuind[0])) #Signal PDF




@jit(nopython=True)
def bgs(psrno, cone, twopi = False):

    '''
    Parameters
    ----------
    psrno : int
        Pulsar number from the ATNF catalogue.
    
    cone : float
        Cone size in degrees.

    twopi : Boolean, optional
        (default) False -> Background PDF is not normalized by 2pi
        True -> Background PDF is normalized by 2pi

    Returns
    -------
    Returns the background PDF for a given pulsar {psrno} within\n
    a cone of {cone} degrees for SINGLE STACKED ANALYSIS
    '''

    if twopi == True:
        s_ang = (np.sin(np.deg2rad(msdec[psrno] + cone)) - np.sin(np.deg2rad(msdec[psrno] - cone)))*(2 * np.pi)
        return 1/s_ang
    else:
        s_ang = (np.sin(np.deg2rad(msdec[psrno] + cone)) - np.sin(np.deg2rad(msdec[psrno] - cone)))
        return 1/s_ang

@jit(nopython=True)
def Pr(x, Ns, S, B):

    '''
    Parameters
    ----------
    x : int
        Assumed no.of associated events

    Ns : int
        No.of neutrinos used for analysis

    S : float
        Signal PDF
    B : float
        Background PDF
    
    Returns
    -------
    float 
        Returns the probability of the selected set of neutrinos being associated\n
        with a given pulsar with {Ns} neutrinos, {S} signal and {B} background PDF and {x} assumed associated events
    '''

    nsN = x/Ns
    return np.add(np.multiply(nsN , S), np.multiply(np.subtract(1, nsN), B))

@jit(nopython=True)
def TS(x, S, B, Ns):

    '''Returns the Test Stastic value for the given value n_s = {x} and its parameters S, B, Ns    '''
    
    return 2*np.sum(np.log(Pr(x,  Ns, S, B)/B))


def ns_for_TSmax(S, B, Ns):

    '''
    Parameters
    ----------
    
    S : float
        Signal PDF
    B : float
        Background PDF
    
    Ns : int
        No.of neutrinos used for analysis
    
    Returns
    -------
    Returns the value of $n_s$ for which
    the TS is maximum for the given Signal,
    Background parameters of a pulsar
    '''

    #returns the TSmax for i^th GRB
    nll = lambda x: -TS(x, S, B, Ns)
    soln = minimize(nll, 6 ,bounds=((0,None),))
    ns = float(soln.x)
    return ns
```

```python
#@jit(nopython=True)
def singpsr(psrno):
    cut = 5
    cone = 5
    S, Ns = sig(psrno, cut)
    B = bgs(psrno, cone, True)
    maxns = ns_for_TSmax(S, B, Ns)
    return [maxns, TS(maxns, S, B, Ns)]

maxts = []
errpsr = []
pool = mul.Pool(8, maxtasksperchild=10)
op_async = pool.map_async(singpsr, range(p))
maxts = op_async.get()
op_async = []
pool = []


maxns = []
sqrtts = []
for i in maxts:
    maxns.append(i[0])
    if abs(i[1]) < 5e-5:
        sqrtts.append(0)
    else:sqrtts.append(i[1])
sqrtts = np.array(sqrtts)
```
